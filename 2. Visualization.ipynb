```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('B. Cleaned_Data.csv' , index_col=0)
```

```python
data_Gender = df["Gender"].value_counts()
data_Geography = df["Geography"].value_counts()
data_HasCrCard = df["HasCrCard"].replace({1:"Yes" , 0:"No"}).value_counts()
data_IsActiveMember = df["IsActiveMember"].replace({1:"Yes" , 0:"No"}).value_counts()

labels_Gender = data_Gender.index
labels_Geography = data_Geography.index
labels_HasCrCard = data_HasCrCard.index
labels_IsActiveMember = data_IsActiveMember.index

palette_color = sns.color_palette('bright') 
fig , ((ax_Gender, ax_Geography) , (ax_HasCrCard , ax_IsActiveMember)) = plt.subplots(ncols=2 , nrows=2)
fig.set_figwidth(10)
fig.set_figheight(10)

ax_Gender.pie(x=data_Gender, labels=labels_Gender , colors=palette_color, autopct='%.1f%%') 
ax_Geography.pie(x=data_Geography, labels=labels_Geography , colors=palette_color, autopct='%.1f%%')
ax_HasCrCard.pie(x=data_HasCrCard, labels=labels_HasCrCard , colors=palette_color, autopct='%.1f%%')
ax_IsActiveMember.pie(x=data_IsActiveMember, labels=labels_IsActiveMember , colors=palette_color, autopct='%.1f%%')

ax_Gender.set_title("Gender")
ax_Geography.set_title("Country")
ax_HasCrCard.set_title("Has Credit Card?")
ax_IsActiveMember.set_title("Is Active Member?")

plt.show()
```

```python
data = df["Balance"].values
sns.histplot(data)
plt.show()
```

```python
zero_balance_count = (df["Balance"] == 0).sum() / df.shape[0] * 100
print(f"\"{zero_balance_count}\" percent of values in the <Balance> are zero.")
```

```python
# this time select non-zero balances
data = df.loc[df["Balance"] != 0 , "Balance"].values
fig , ax = plt.subplots(nrows=1 , ncols=1)
ax = sns.histplot(data , kde=True)
ax.set_title("<Balance> Distribution")
fig.set_figheight(10)
fig.set_figwidth(20)
plt.show()
```

```python
# this time select non-zero balances
data = df["CreditScore"].values
fig , ax = plt.subplots(nrows=1 , ncols=1)
ax = sns.histplot(data , kde=True)
ax.set_title("<CreditScore> Distribution")
fig.set_figheight(10)
fig.set_figwidth(20)
plt.show()
```

```python
data = df["Tenure"].value_counts().sort_index()
labels = data.index
palette_color = sns.color_palette('dark') 

fig , ax = plt.subplots(ncols=1 , nrows=1)
fig.set_figwidth(7)
fig.set_figheight(7)

ax.pie(x=data, labels=labels , colors=palette_color, autopct='%.1f%%') 
ax.set_title("Tenure Distribution")
plt.show()
```
